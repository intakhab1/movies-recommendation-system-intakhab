```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# Stemming is required SO import NLTK library
!pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
 new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Now we calculate angle between vectors ( here we have 5000 dimensional vector)
# we calculate cosine distance between two vectors 
# Usingcosine similarity function scikit learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector).shape
```

```python
similarity = cosine_similarity(vector)
```
